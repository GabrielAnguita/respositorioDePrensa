```python

import requests
from bs4 import BeautifulSoup as bs
import json

import re


    
def dict_noticia(noticia):  
    cleanr = re.compile('{.*?}')
    texto = bs(noticia["_source"]["texto"], "html.parser").text
    textolimpio = re.sub(cleanr, '', texto)  
    
    noti = {}
    noti["id"] = noticia["_id"]
    noti["autor"] = noticia["_source"]["autor"]
    noti["fuente"] = noticia["_source"]["fuente"]
    noti["seccion"] = noticia["_source"]["seccion"]
    noti["subSeccion"] = noticia["_source"]["subSeccion"]
    noti["fechaModificacion"] = noticia["_source"]["fechaModificacion"]
    noti["fechaPublicacion"] = noticia["_source"]["fechaPublicacion"]
    noti["temas"] = noticia["_source"]["temas"]
    noti["titulo"] = noticia["_source"]["titulo"]
    noti["bajada"] = noticia["_source"]["bajada"][0]["texto"]
    noti["texto"] = textolimpio
    noti["url"] = noticia["_source"]["permalink"]
    noti["tituloSEO"] = noticia["_source"]["tituloSEO"]
    noti["relevancia"] = noticia["_source"]["relevancia"]
    
    return noti
```

```python
import time

headers = { "Host": "cache-elastic-pandora.ecn.cl",
"User-Agent": "Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:75.0) Gecko/20100101 Firefox/75.0",
"Accept": "application/json, text/javascript, */*; q=0.01",
"Accept-Language": "en-US,en;q=0.5",
"Accept-Encoding": "gzip, deflate, br",
"Origin": "https://www.emol.com",
"Connection": "keep-alive",
"Referer": "https://www.emol.com/todas/" }

url = "https://cache-elastic-pandora.ecn.cl/emol/noticia/_search?q=publicada:true+AND+ultimoMinuto:true+AND+seccion:*+AND+temas.id:*&sort=fechaModificacion:desc&size=30&from=0"


lista_noticias = []

for i in range(20):
    link = url + str(i * 30)

    response = requests.get(link, headers=headers)

    payload = json.loads(response.text)

    for noticia in payload["hits"]["hits"]:
        lista_noticias.append(dict_noticia(noticia))
        
    time.sleep(2)
```

```python
def contar(tag):
    cuenta = 0
    for lista in todo_tag:
        if tag in lista:
            cuenta += 1
    return cuenta
```

```python
tags = {}
todo_tag = [[] for x in range(len(lista_noticias))]

for idx, noticia in enumerate(lista_noticias):
    for tema in noticia["temas"]:
        if tema["nombre"] not in tags.keys():
            tags.update({tema["nombre"]:0})
        todo_tag[idx].append(tema["nombre"])

for tag in tags.keys():
    tags[tag] = contar(tag)
    
#for noticia in lista_noticias:
#    lista_noticias[0].keys()
```
